```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd 
import seaborn as sns
# Keras
from keras.models import Sequential
from keras.layers import Dense

from tensorflow.keras.optimizers import SGD, Adam, Adadelta, RMSprop
import keras.backend as K
# Train-Test
from sklearn.model_selection import train_test_split
# Scaling data
from sklearn.preprocessing import StandardScaler
# Classification Report
from sklearn.metrics import classification_report
from keras.utils.np_utils import to_categorical
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df = df.T
```

```python
new_df = df.drop(['group','samples'])
new_df[2] = (new_df[2] +new_df[3])/2
new_df[4] = (new_df[4] +new_df[5])/2
new_df = new_df.drop([3, 5], axis=1)
new_df.head()
```

```python
new_df.rename(columns = {0:'group_0', 1:'group_1',
                              2:'group_2',4:'group_3'}, inplace = True)
new_df.head()
```

```python
metabolites = new_df.index
metabolites = list(metabolites)
Metabolites = [4*metabolites]
#print(Metabolites[0])
group_0 = list(new_df['group_0'])
group_1 = list(new_df['group_1'])
group_2 = list(new_df['group_2'])
group_3 = list(new_df['group_3'])
concentration = []
concentration.extend(group_0)
concentration.extend(group_1)
concentration.extend(group_2)
concentration.extend(group_3)

def group_label(group,val):
    label = []
    for i in range(0,len(group)):
        label.append(val)
    return label
group = []
group = group_label(group_0,0)
group.extend(group_label(group_1,1))
group.extend(group_label(group_2,2))
group.extend(group_label(group_3,3))
group
```

```python
concn = [round(item) for item in concentration ]
print(concn)
```

```python
def value_10prcnt(value):
    bound_value = round(0.1 * value)
    lower_bound = value - bound_value
    upper_bound = value + bound_value
    value_range = []
    for i in range(lower_bound,(upper_bound+1)):
        value_range.append(i)
    return value_range
```

```python
def value_range_generator(value,length):
    value_range = []
    for i in range(0,length):
        value_range.append(value)
    return value_range
```

```python
concentration_range = []
Metabolites_range = []
group_label = []


for i in range(0,len(concn)):
    concn_range = value_10prcnt(concn[i])
    temp_length = len(concn_range)
    metabolite_range = value_range_generator(Metabolites[0][i],temp_length)
    group_range = value_range_generator(group[i],temp_length)
    concentration_range.extend(concn_range)
    Metabolites_range.extend(metabolite_range)
    group_label.extend(group_range)
    

print(len(concentration_range))
print(len(Metabolites_range))
print(len(group_label))
```

```python
dataset = pd.DataFrame(list(zip(Metabolites_range, concentration_range)),
               columns =['Metabolites', 'Concentration'])
```

```python
dataset['Group']=group_label
```

```python
print(dataset.head())
print(dataset.shape)
```

```python
# saving the dataframe
#dataset.to_csv('processed_data.csv')
```

Model Training

```python
metabolites_cat = pd.get_dummies(dataset['Metabolites'])
```

```python
metabolites_cat
```

```python
new_dataset = pd.concat([dataset,metabolites_cat],axis=1)
```

```python
new_dataset=new_dataset.drop(['Metabolites'],axis=1)
```

```python
new_dataset
```

```python
X = new_dataset.drop(['Group'],axis = 1)
y = new_dataset['Group']
y
```

```python
X_scaled = X.copy()
X_scaled["Concentration"] = X_scaled["Concentration"] /X_scaled["Concentration"].abs().max()
list(X_scaled.iloc[0,:])
```

```python
y_cat = to_categorical(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_cat, test_size=0.2)
```

```python
# Feature Scaling
#from sklearn.preprocessing import StandardScaler
#sc = StandardScaler()
#X_train = sc.fit_transform(X_train)
#X_test = sc.transform(X_test)
```

```python
model = Sequential()
model.add(Dense(60, input_shape = (54,), activation = "relu"))
model.add(Dense(15, activation = "relu"))
#model.add(Dropout(0.2))
model.add(Dense(4, activation = "softmax"))
model.compile(Adam(lr = 0.01), "categorical_crossentropy", metrics = ["accuracy"])
model.summary()
```

```python
model.fit(X_train, y_train, verbose=1, epochs=100)
```

```python
y_predict = np.argmax(model.predict(X_test), axis=-1)
```
